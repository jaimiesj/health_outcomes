```python
import pandas as pd
```

### Load Raw Data 

```python
raw_data_path = '../res/Student Alcohol Consumption Merged.csv'
raw_df = pd.read_csv(raw_data_path)
raw_df.head()
```

### Select useful columns as features for the model

```python
features = ['school', 'sex', 'age', 'famsize', 'studytime', 'famsup', 'activities', 'internet', 'romantic', 'famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health', 'absences']

filtered_df = raw_df[features]
filtered_df.head()
```

### Rename the columns with more descriptive titles

```python
old_names = filtered_df.columns

names_dict = {
    'Medu': 'fathers education', 'Fedu': 'mothers education', 'Dalc': 'weekday drinker', 'Walc': 'weekend drinker',
    'famsup': 'financial support', 'goout': 'social activities', 'romantic': 'significant other', 'freetime': 'free time',
    'famrel': 'family relationship', 'studytime': 'study time', 'famsize': 'household size', 'sex': 'gender'
} 

replace = lambda entry: entry if entry not in names_dict else names_dict[entry]

new_names = pd.Series(map(replace, old_names))
filtered_df.columns = new_names
filtered_df.head()
```

### Create a csv file from the filtered dataframe

```python
filtered_data_path = '../res/alcoholism_dataset.csv'
filtered_df.to_csv(filtered_data_path, index=False)
```
